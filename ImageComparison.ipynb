**Image comparison**

Comparing images on basis of fidelity and diversity is difficult task.We should be clear on what we want to compare.There are two approaches for this task.
Pixel to pixel approach may fail in some scenarios or is not just sufficient for comparison.There is second approach called feature distance which will give better reliability in comparison.
Pixel distance is simple but not much reliable. Feature distance uses higher level features of images making it more reliable.
